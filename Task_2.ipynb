# Customer Support Ticket data analysis

```python
#imports for Customer Support Ticket Dataset and analysing it
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading data
data = pd.read_csv('customer_support_tickets.csv')
data.head()
```

```python
#imports for clean and preprocess data
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import string
from collections import Counter
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
#Define function for cleaning text
def clean_text(text):
  #Convert to lower_case
  text = text.lower()
  #Remove punctuation

  text = text.translate(str.maketrans('','',string.punctuation))
  #Remove stopword and tokenize
  stop_words = set(stopwords.words('english'))
  word_tokens = word_tokenize(text)
  filtered_text = [word for word in word_tokens if word not in stop_words]
  return ' '.join(filtered_text)

#Use function to clean given data
data["cleaned_text"] = data['Ticket Description'].apply(clean_text)
data.head()
```

```python
#import counter and 10 most common keywords used and print it
common_words = Counter(' '.join(data['cleaned_text']).split()).most_common(10)
print(common_words)
```

```python
#Calculate category counts for each type of problems in categories
category_counts = data['Ticket Subject'].value_counts()
print(category_counts)
```

```python
#Ticket Resolution / Response Times
data['Time to Resolution'] = pd.to_datetime(data['Time to Resolution'])
response_times = data['Time to Resolution'].mean()
print(f"Time Taken to Respond: ", response_times)
```

```python
#Sentiment Analysis
sid = SentimentIntensityAnalyzer()

data['sentiment_score'] = data['cleaned_text'].apply(lambda x: sid.polarity_scores(x)['compound'])

#Clustering
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(data['cleaned_text'])
kmeans = KMeans(n_clusters=5,random_state = 1234)
data['cluster'] = kmeans.fit_predict(X)

```

```python
data.head()
```

```python
def segregation(sentiment_score):
     if sentiment_score>0.05:
        return 'positive'
     elif sentiment_score<0.05:
        return 'negative'
     else:
        return 'neutral'

data['sentiment'] = data['sentiment_score'].apply(segregation)
data.head()
```
